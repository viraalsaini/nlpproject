```python
def translate_en_to_it(text):
    tokenized_text = tokenizer_en_to_it(text, return_tensors="pt", padding=True)
    translated = model_en_to_it.generate(**tokenized_text)
    return tokenizer_en_to_it.decode(translated[0], skip_special_tokens=True)
```

```python
import nltk
from nltk.util import ngrams
nltk.download('punkt')
def generate_ngrams(text, n=2):
    words = nltk.word_tokenize(text)
    n_grams = list(ngrams(words, n))
    return [" ".join(gram) for gram in n_grams]
text = "Buongiorno, come stai oggi?"
print(generate_ngrams(text, 2))  # ['Buongiorno, come', 'come stai', 'stai oggi'
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
def pos_tagging(text):
    doc = nlp(text)
    return [(token.text, token.pos_) for token in doc]
text = "I am learning NLP."
print(pos_tagging(text))
```

```python
def named_entity_recognition(text):
    doc = nlp(text)
    return [(ent.text, ent.label_) for ent in doc.ents]
text = "Barack Obama è stato presidente degli Stati Uniti."
print(named_entity_recognition(text))
```

```python
from transformers import pipeline

sentiment_pipeline = pipeline("sentiment-analysis")

try:
    hate_speech_pipeline = pipeline("text-classification", model="unitary/toxicity")
except:
    print("Hate speech detection pipeline not found. Please check if the model is available.")
    hate_speech_pipeline = None

try:
    sarcasm_pipeline = pipeline("text-classification", model="bhadresh-savani/bert-base-uncased-sarcasm")
except:
    print("Sarcasm detection pipeline not found. Please check if the model is available.")
    sarcasm_pipeline = None

def analyze_sentiment_and_hate(text):
    results = {}

    sentiment_result = sentiment_pipeline(text)
    results["sentiment"] = sentiment_result[0]

    if hate_speech_pipeline:
        hate_speech_result = hate_speech_pipeline(text)
        results["hate_speech"] = hate_speech_result[0]

    if sarcasm_pipeline:
        sarcasm_result = sarcasm_pipeline(text)
        results["sarcasm"] = sarcasm_result[0]

    return results

text = "I hate you, you stupid idiot."
analysis = analyze_sentiment_and_hate(text)
print(analysis)
```
